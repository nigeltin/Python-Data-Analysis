```python
sales = pd.read_csv('salesmen.csv', parse_dates=['Date'])
sales['Salesman'] = sales['Salesman'].astype('category')
sales.head(3)
```

```python
sales.info()
```

```python
sales['Salesman'].nunique()
```

```python
#pivot() method
sales.pivot(index='Date', columns = 'Salesman', values='Revenue')
```

```python
#pivot_table method
foods = pd.read_csv('foods.csv')
foods.head(3)
```

```python
foods.pivot_table(values='Spend', index=['Gender', 'Item'], columns='City', aggfunc='mean')
```

```python
#pd.melt works in reverse to the pivot method

quart = pd.read_csv('quarters.csv')
quart.head(3)
```

```python
pd.melt(quart, id_vars='Salesman')
```
